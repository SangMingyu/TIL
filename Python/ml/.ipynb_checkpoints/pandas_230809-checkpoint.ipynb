```python
df_boston = pd.read_csv('./data/boston.csv')
df_boston
```

```python
## 열 및 행선택
- CRIM, ZN, INDUS
```

```python
result = df_boston[['CRIM', 'ZN', 'INDUS']]
result
cols = ['CRIM', 'ZN', 'INDUS']
result = df_boston[cols]
```

```python
- ZN이 18.0에 해당하는 행만 추출
# 행 추출의 기본 원리는 조건문
```

```python
result = df_boston[df_boston['ZN'] == 18.0]
result
```

```python
- AGE가 50이상인 데이터만 추출
```

```python
result = df_boston[df_boston['AGE'] >= 50].reset_index(drop=True)
result
```

```python
- INDUS가 5.0 이상이면서 B가 394이상인 데이터 추출
```

```python
result = df_boston[(조건식1) & (조건식2)]
result = df_boston[(df_boston['INDUS'] >= 5.0) & (df_boston['B'] >= 394)].reset_index(drop=True)
result
```

```python
## iloc와 loc 문법
```

```python
df_boston.head()
```

```python
# iloc 활용해서 추출
result = df_boston.iloc[0:3, 0:3]
result
```

```python
# loc 활용해서 추출
result = df_boston.loc[0:2, ['CRIM', 'ZN', 'INDUS']]
result
```

```python
result = df_boston.loc[(df_boston['INDUS'] >= 5.0) & (df_boston['B'] >= 394) # 행
                     ,['CRIM', 'ZN', 'INDUS']].reset_index(drop=True) # 열
result
```

```python
import pandas as pd 
import numpy as np

np.random.seed(0) # 초전도체 현상, 실험조건 동일하게 맞추기 위해
data = np.random.randint(0, 10, size=(10, 3)).astype(float)
print(data)
```

```python
data[data < 3] = np.nan
print(data)
```

```python
df = pd.DataFrame(data,columns=["col1", "col2", "col3"]) # 외부에서 새로운 데이터 가져옴 

df.info()
```
