```python
import pandas as pd 
import numpy as np

np.random.seed(0) # 초전도체 현상, 실험조건 동일하게 맞추기 위해
data = np.random.randint(0, 10, size=(10, 3)).astype(float)
print(data)
```

```python
data[data < 3] = np.nan
print(data)
```

```python
df = pd.DataFrame(data,columns=["col1", "col2", "col3"]) # 외부에서 새로운 데이터 가져옴 

df.info()
```

```python
df.isna().sum()
```

```python
# 평균으로 대치 해보기
df['col2'].fillna(int(df['col2'].mean(axis=0)), inplace = True)
df['col3'].fillna(int(df['col3'].mean(axis=0)), inplace = True)
df
```

```python
avg = df.mean()
df.fillna(avg, inplace=True)
df # 이런방법도 있다.
```

```python
col3_mean = df['col3'].mean()
col3_mean
df['col3'] = df['col3'].fillna(col3_mean)
df['col2'] = df['col2'].fillna(df['col2'].mean())
df
```

```python
df = pd.DataFrame({
    "이름": ["Evan", "Sara", "Chris", "Messi", "Woods", "Jordan"], 
    "국어": [70, 90, None, None, 100, 85], 
    "영어": [65, 100, None, 90, 70, 80]
})

df
```

```python
## 행 추가
df.loc[6, "이름":"영어"] = ["Kim", 50, 60]
df
```

```python
# 국어점수가 90점 이상이면 "A"
# 국어점수가 80점~89점이면 "B"
# 국어점수가 70점~79점이면 "C"
# 나머지는 F

condition_list = [(df["국어"] >= 90), (df['국어'] >= 80) & (df['국어'] < 90), (df['국어'] >= 70) & (df['국어'] < 80)]
choice_list = ["A", "B", "C"]
df['점수'] = np.select(condition_list, choice_list, default = "F")
df
```

```python
- numpy.where, 조건식이 한번만 적용
=if 조건식, 참일 때, 거짓일 때
```

```python
# np.where(조건식, 참, np.where(조건식, 참, (np.where(조건식, 참)))))
# np.select(다중조건식, 다중결과값, default==> else 조건에 해당)
```

```python
covidtotalsonly = covidtotals.loc[:, totvars]
covidtotalsonly['total_cases_q'] = pd.qcut(covidtotalsonly['total_cases'],
                                          labels = ['매우낮음', '낮음', '중간', '높음', '매우높음'],
                                          q=5, precision=0)

covidtotalsonly['total_deaths_q'] = pd.qcut(covidtotalsonly['total_deaths'],
                                          labels = ['매우낮음', '낮음', '중간', '높음', '매우높음'],
                                          q=5, precision=0)

covidtotalsonly.head(1)
```

```python
pd.crosstab(covidtotalsonly['total_cases_q'], covidtotalsonly['total_deaths_q'])
```
