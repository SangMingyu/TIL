```python
## 미션
#- JSON 데이터를 pandas 데이터프레임으로 가공

import pandas as pd
import json

cars = json_df['realUnitTrtmVO']
pd.DataFrame(cars)
```

```python
records = []
for car in cars:
    row_dict = {}
    row_dict["date"] = car["stdDate"]
    row_dict["timeAvg"] = car["timeAvg"]
    records.append(row_dict)
pd.DataFrame(records)
```

```python

## XML테스트
import requests
import lxml
from bs4 import BeautifulSoup
import pandas as pd 

url = "http://data.ex.co.kr/openapi/trtm/realUnitTrtm?key=1308370912&type=xml&iStartUnitCode=101&iEndUnitCode=103&numOfRows=10&pageNo=1"
req = requests.get(url)
#print(req.text)
soup = BeautifulSoup(req.text, "lxml") # XML 반환값이라서 LXML을 사용한 것
#print(soup.prettify())

stddate_list = soup.find_all("stddate")
timeAvg_list = soup.find_all("timeavg")

stddate = []
avgTime = []
for date, timeAvg in zip(stddate_list, timeAvg_list):
    #print(date.get_text(), timeAvg.get_text())
    stddate.append(date.get_text())
    avgTime.append(timeAvg.get_text())

df = pd.DataFrame({"date":stddate, "avgTime":avgTime})
df
```

```python

import requests
import lxml
from bs4 import BeautifulSoup
import pandas as pd 

url = "http://openapi.seoul.go.kr:8088/4e46586a6d73736138376451547649/xml/tbLnOpendataRtmsV/1/5/"
req = requests.get(url)
#print(req.text)
soup = BeautifulSoup(req.text, "lxml")
#print(soup.prettify())

sgg = soup.find_all("sgg_nm")
price = soup.find_all("obj_amt")

sgg_list = []
price_list = []
for sgg_nm, obj_amt in zip(sgg, price):
    #print(sgg_nm.get_text(), obj_amt.get_text())
    sgg_list.append(sgg_nm.get_text())
    price_list.append(obj_amt.get_text())

df = pd.DataFrame({"자치구명":sgg_list, "금액":price_list})
df
```
